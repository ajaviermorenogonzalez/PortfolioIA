# Machine Learning - Predicción diabetes

## Datos y definición del problema


Dataset:              https://www.kaggle.com/datasets/iammustafatz/diabetes-prediction-dataset

Datos: 
    
    gender:               Genero del paciente: Male, Female, Other
    age:                  Edad del paciente
    hypertension:         Si el paciente padece (1) o no (0) hipertensión
    heart_disease:        Si el paciente padece (1) o no (0) enfermedad de corazón
    smoking_history:      Historial fumador del paciente ['never', 'former', 'not current', 'ever', 'current', 'No Info']
    bmi:                  Indice de masa corporal del paciente
    HbA1c_level:          Nivel de hemoglobina glicosilada (HbA1c)
    blood_glucose_level:  Nivel de glucosa en sangre
    diabetes:             Si el paciente padece (1) o no (0) diabetes

Problema:             Predecir variable objetivo diabetes de tipo binaria


## Imports y Set up

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import joblib

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder as label_encoder
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score, learning_curve
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, roc_curve, confusion_matrix
from sklearn.impute import KNNImputer
from tqdm import tqdm

# Muestra todas las columnas por pantalla
pd.set_option('display.max_columns', None)

# Crea una instancia del escalador estándar
scaler = StandardScaler()
encoder = label_encoder()
```

## Preprocesamiento de datos

```python
"""
    Gender: Categorico 3 tipos
        - Codifico con getDummies 
        - deberia usar drop_first = true para evitarme una columna pero se leerian peor los datos
        
    smoking_history: Categorico 5 tipos
        - Codifico con label encoder 
        - Imputar valores 'No Info' infiriendo que otro tipo sería a partir del resto de datos
    
"""
def preprocesamiento(data):
    encoder = label_encoder()
    
    # GENDER

    # Obtener las variables dummies y añadirlas al dataframe original 
    dummies_gender = pd.get_dummies(data['gender'])
    data = pd.concat([data, dummies_gender], axis=1)
    data = data.drop('gender', axis=1)

    # SMOKING_HISTORY

    # Crear una copia del DataFrame para realizar la inferencia
    data_aux = data.copy()

    # Orden deseado
    orden = ['never', 'former', 'not current', 'ever', 'current', 'No Info']

    # Indico mi orden al encoder
    encoder = encoder.fit(orden)
    data_aux['smoking_history'] = encoder.transform(data_aux['smoking_history'])

    # Reemplazar 'No Info' por NaN en la columna 'smoking_history'
    data_aux['smoking_history'].replace(0, np.nan, inplace=True)
    
    # Procesamiento valores nulos 

    # Crear instancia del KNN Imputer con k=5
    imputer = KNNImputer(n_neighbors=5)

    # Realizar la inferencia de los valores faltantes 
    data_imputed_values = imputer.fit_transform(data_aux)

    # Convertir los valores inferidos en un DataFrame
    data_imputed_values = pd.DataFrame(data_imputed_values, columns=data_aux.columns)

    # Redondeo a entero
    data_imputed_values = data_imputed_values['smoking_history'].round(0)

    # Reemplazar los valores 'No Info' en el DataFrame original con los valores inferidos
    #df['smoking_history'] = df_imputed_values
    data['smoking_history']  = encoder.inverse_transform(data_imputed_values.array.astype(int))
    print(data['smoking_history'].value_counts())
    
    # Aplicamos label encoder a smoking_history
    data['smoking_history'] = encoder.fit_transform(data['smoking_history'])

    # Aplicamos escalado estandar a Edad y Glucosa
    
    print("\nEstandarizamos Edad y Glucosa")
    data['age'] = scaler.fit_transform(data['age'].array.reshape(-1, 1))
    data['blood_glucose_level'] = scaler.fit_transform(data['blood_glucose_level'].array.reshape(-1, 1))
    data['bmi'] = scaler.fit_transform(data['bmi'].array.reshape(-1, 1))
    
    # Validacion 
    print("\nComprobamos los datos procesados")
    visualizacion(data)
    
    return data 
```

## Configuramos el modelo

```python
# Dividimos nuestros datos en datos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 420)

# Vamos a utilizar el modelo RandomForest
rf = RandomForestClassifier()

# Numero de arboles de decision
n_estimators = [int(x) for x in np.linspace(start = 1000, stop = 1200, num = 12)]

# Numero maximo de caracteristicas a usar por el arbol
max_features = ['auto', 'sqrt']

# Maxima profundidad del arbol
max_depth = [int(x) for x in np.linspace(start = 5, stop = 30, num = 6)]

# Numero de muestras necesarios para tomar el conjunto de datos que se necesita
min_samples_split = [2, 5, 10, 15, 100]

# Minimo numero de muestras por hoja/nodo
min_samples_leaf = [1, 2, 5, 10]

# Diccionario que toma todas las caracteristicas anteriores
random_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'max_depth': max_depth,
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf
    }

print("Random Grid:", random_grid)


```

## Entrenamos el modelo

```python

# Este entrenamiento es muy pesado ya que busca las mejores características con validación cruzada y un gran número de
# parámetros. Almacenamos el modelo entrenado para cargarlo posteriormente.

"""
# Realiza la búsqueda aleatoria con tqdm para mostrar el progreso
progress_bar = tqdm(total=10, desc="Hyperparameter Search", ncols=100)
for _ in range(10):
    final_model = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, scoring='roc_auc',
                                     n_iter=1, cv=5, random_state=42, n_jobs=4, verbose=0)

    final_model.fit(X_train, y_train)

    # Actualiza la barra de progreso
    progress_bar.update(1)

progress_bar.close()

joblib.dump(final_model, 'model.joblib')
"""
```

## Evaluamos el modelo

```python
# Realizamos la prediccion y evaluamos su resultado
y_pred = model.predict(X_test)
y_pred_binary = np.where(y_pred > 0.5, 1, 0)
```

### Calculamos el área bajo la curva

```python
# Calcular el área bajo la curva ROC
roc_auc = roc_auc_score(y_test, y_pred, average='macro')

# Calcular la curva ROC
fpr, tpr, _ = roc_curve(y_test.ravel(), y_pred.ravel())

# Crear el gráfico de la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='Curva ROC (AUC = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.savefig('plots/area_bajo_curva.png')
plt.show()
```

```python
# Obtenemos sus metricas
print('Area bajo la curva (AUC-ROC)')

# Calcula el AUC-ROC
auc_roc = roc_auc_score(y_test, y_pred)

print(auc_roc)  
```

#### Un area bajo la curva de 0.84 indica un buen desempeño del modelo

### Comparacion valores reales y predichos

```python
# Calcula la precisión del modelo comparando las etiquetas predichas con las etiquetas reales
accuracy = accuracy_score(y_test, y_pred_binary)
print("Precisión:")
print(accuracy)
```

### Construimos su matriz de confusión

```python
# Calcular la matriz de confusión
confusion = confusion_matrix(y_test, y_pred_binary)

labels = ["No Diabetes", "Diabetes"]

# Imprimir la matriz de confusión como un heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicciones')
plt.ylabel('Etiquetas verdaderas')
plt.title('Matriz de Confusión')
plt.savefig('plots/matriz_confusion.png')
plt.show()
```

#### La gran diferencia entre valores predicciones erroneas puede indicar algo de sobreentrenamiento

### Comprobamos un posible overfitting estudiando las curvas de entrenamiento 

```python
# Cargamos el modelo y sus puntuaciones
model_overfitting = joblib.load('model_overfitting.joblib')

def plot_learning_curve(estimator, X, y, cv, train_sizes):
    train_scores = []
    test_scores = []

    with tqdm(total=len(train_sizes), desc="Calculando curvas de aprendizaje") as pbar:
        for train_size in train_sizes:
            _, train_score, test_score = learning_curve(estimator, X, y, cv=cv, train_sizes=[train_size], scoring='accuracy', n_jobs=-1)
            train_scores.append(np.mean(train_score))
            test_scores.append(np.mean(test_score))
            pbar.update(1)

    plt.figure()
    plt.title('Curvas de Aprendizaje')
    plt.xlabel('Tamaño del conjunto de entrenamiento')
    plt.ylabel('Precisión')
    plt.grid()

    plt.plot(train_sizes, train_scores, 'o-', color='r', label='Precisión entrenamiento')
    plt.plot(train_sizes, test_scores, 'o-', color='g', label='Precisión validación')

    plt.legend(loc='best')
    
    plt.savefig('plots/curvas_aprendizaje.png')

    plt.show()

train_sizes = np.linspace(0.1, 1.0, 10)  

# Este proceso puede tardar un poco por lo que ya lo hemos calculado y almacenado para mostrarlo 
# plot_learning_curve(model_overfitting, X, y, cv=5, train_sizes=train_sizes)
```
